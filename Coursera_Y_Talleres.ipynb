```python
import matplotlib.pyplot as plt
import numpy as np

# --- Función para generar degradé entre dos colores ---
def generar_degrade_hex(hex1, hex2, n):
    c1 = np.array([int(hex1[i:i+2], 16) for i in (1,3,5)])
    c2 = np.array([int(hex2[i:i+2], 16) for i in (1,3,5)])
    colores = [(c1 + (c2 - c1) * t).astype(int) for t in np.linspace(0, 1, n)]
    return [f"#{r:02x}{g:02x}{b:02x}" for r, g, b in colores]

# --- Datos ---
cursos = [
    "Realidad Extendida y Metaverso para todos", 
    "Desarrollo XR en Unity para todos",
    "Creando proyectos XR y Metaversos",
    "Desarrollo de Metaversos para todos"
]

visitantes = [4407, 1769, 753, 704]

# --- Colores (4 colores en degradé XR) ---
colores = generar_degrade_hex("#8c52ff", "#5ce1e6", len(cursos))

plt.figure(figsize=(10,6))
barras = plt.bar(cursos, visitantes, color=colores)

# --- Título y etiquetas ---
plt.title("Visitantes por Curso XR")
plt.ylabel("Número de visitantes")
plt.xticks([])   # Oculta los nombres abajo


# --- Etiquetas encima de cada barra ---
for i, v in enumerate(visitantes):
    plt.text(i, v + max(visitantes)*0.02, str(v), ha='center', va='bottom')

# --- Leyenda personalizada ---
handles = [plt.Rectangle((0,0),1,1, color=colores[i]) for i in range(len(cursos))]
plt.legend(handles, cursos, title="Cursos", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

En los analisis siguientes se realizo el calculo de cuanta gente en verdad contesto las encuestas de estas  variables.

```python
def analizar_columna(data, columna, nombre_mostrar, traduccion=None):
    # Filtrar quienes reportaron algo
    data_filtrado = data[data[columna].notna() & (data[columna] != '')]
    
    total = len(data)
    respondieron = len(data_filtrado)
    porcentaje_respondieron = (respondieron / total) * 100
    
    # Distribución (porcentaje entre quienes respondieron)
    dist = data_filtrado[columna].value_counts(normalize=True) * 100
    
    # Mostrar resultados en texto
    print(f"\n{porcentaje_respondieron:.1f}% de los estudiantes reportaron su {nombre_mostrar.lower()}.")
    
    for valor, porcentaje in dist.items():
        valor_mostrar = traduccion.get(valor, valor) if traduccion else valor
        print(f"De ese porcentaje, {porcentaje:.1f}% corresponden a {valor_mostrar.lower()}.")
    
    # --- Gráfica ---
    etiquetas = [traduccion.get(v, v) if traduccion else v for v in dist.index]
    plt.figure(figsize=(6, 4))
    plt.bar(etiquetas, dist.values)
    plt.title(f"Distribución de {nombre_mostrar}")
    plt.ylabel("Porcentaje (%)")
    plt.xticks(rotation=45, ha='right')
    plt.show()

# --- 1. Género ---
traduccion_genero = {'male': 'Masculino', 'female': 'Femenino'}
analizar_columna(data, 'reported_or_inferred_gender', 'Género', traduccion_genero)

# --- 2. Situación laboral ---
analizar_columna(data, 'employment_status', 'Situación laboral')

# --- 3. Situación de estudio ---
analizar_columna(data, 'student_status', 'Situación de estudio')
```
